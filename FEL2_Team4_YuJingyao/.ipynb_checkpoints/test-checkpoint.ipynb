```python
# 定义学生对象
class Student:
    def __init__(self,tutorial_group, student_id, school, name, gender, cgpa):
        self.tutotial_group = tutorial_group
        self.student_id = student_id
        self.school = school
        self.name = name
        self.gender = gender
        self.cgpa = cgpa
    def __lt__(self, other):
        if not isinstance(other, Student):
            return NotImplemented
        return self.cgpa < other.cgpa
    def __eq__(self, other):
        if not isinstance(other, Student):
            return NotImplemented
        return self.cgpa == other.cgpa
    def __repr__(self):
        return f"\nName: {self.name}; Gender:{self.gender} ;School: {self.school}; CGPA: {self.cgpa}"
st1 = Student('GP-1',111,"EEE","Zhang San",'male',4.3)
st2 = Student('GP-6',123,"SSS","Li Si",'male',4.0)
st3 = Student('GP-45',777,"MAE","Wang Wu",'male',4.5)
print(st1)
print((st1.cgpa+st2.cgpa)/2)
st_gp = [st1,st2,st3]
print(sorted(st_gp))
```

```python
# 读取数据，创建学生对象
def read_data(file_path:str):
    with open(file_path,'r') as f:
        f.readline()
        dataset={}
        for line in f:
            tutorial_group, student_id, school, name, gender, cgpa = line.strip().split(",")
            student_obj = Student(
                tutorial_group = tutorial_group,
                student_id = student_id,
                school = school,
                name = name,
                gender = gender,
                cgpa = float(cgpa)
                )
            if tutorial_group not in dataset.keys():
                dataset[tutorial_group] = []
            dataset[tutorial_group].append(student_obj)
    return dataset
```

```python
# 评分标准，这里性别>院校>gpa
def grading(cur_gp:dict,new_member:Student,point):
    # gender
    if cur_gp["male_count"] >= 3 or cur_gp["female_count"] >= 3:
        point -= 2500
    else:
        point += 50
    # school
    if new_member.school in cur_gp["school"]:
        school_count = 0
        for member in cur_gp["member"]:
            if member.school == new_member.school:
                school_count += 1
        if school_count >= 2: 
            point -= 1000 
        else:
            point -= 20
    else:
        point += 35
    
    # gpa
    if len(cur_gp["member"]) > 0:
        avg_gpa = sum(i.cgpa for i in cur_gp["member"]) / len(cur_gp["member"])
    # 奖励与当前队伍平均分差异大的候选人
        cgpa_bonus = abs(avg_gpa - new_member.cgpa) * 12
        point += cgpa_bonus
    return point

# 选出一个5人小组
def select(gp:list):
    cur_gp = {
        'male_count': 0,
        'female_count': 0,
        'gp_avg': sum(i.cgpa for i in gp)/len(gp),
        'school': [],
        'member': []
    }

    first_member = gp.pop(len(gp)//2)
    if first_member.gender == 'Male':
        cur_gp['male_count'] += 1
    else:
        cur_gp['female_count'] += 1
    if first_member.school not in cur_gp["school"]:
        cur_gp["school"].append(first_member.school)
    cur_gp["member"].append(first_member)
    
    for i in range(4):
        max_point = -99999
        for index,member in enumerate(gp):
            point = grading(cur_gp,member,0)
            if point > max_point:
                max_point = point
                best_member = member
                best_index = index
        if best_member.gender == 'Male':
            cur_gp['male_count'] += 1
        else:
            cur_gp['female_count'] += 1
        if best_member.school not in cur_gp["school"]:
            cur_gp["school"].append(best_member.school)
        cur_gp["member"].append(best_member)
        gp.pop(best_index)
    return cur_gp

# 对一个大组里50人分组
def distribution(gp:list):
    distribut = {i:[] for i in range(10)}
    standard = [[] for i in range(10)]
    for i in range(10):
        cur_gp = select(gp)
        distribut[i] = cur_gp["member"]
        if cur_gp["male_count"] <= 3 and cur_gp["female_count"] <= 3:
            standard[i].append(1)
        else:
            standard[i].append(0)
        standard[i].append(len(cur_gp["school"]))
        standard[i].append(round(sum(i.cgpa for i in cur_gp["member"])/5,2))
    return distribut,standard
```

```python
# 对全部120个大组进行分组
def distribut_all(df:dict):
    arrangement = {i:{} for i in df.keys()}
    standard = [[] for _ in range(len(df.keys()))]
    count = 0
    for gp_name, gp in df.items():
        df[gp_name] = sorted(gp)
        arrangement[gp_name],standard[count]= distribution(gp)
        count += 1
    return arrangement,standard
```

```python
df = read_data('records.csv')

# 分类结果
res,std = distribut_all(df)
print(res)
```
